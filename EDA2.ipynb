```python
import kagglehub


path = kagglehub.dataset_download("vivek468/superstore-dataset-final")

print("Path to dataset files:", path)
```

```python
import os
import pandas as pd
print(os.listdir(path))

csv_path = os.path.join(path, "Sample - Superstore.csv")
df_superstore = pd.read_csv(csv_path, encoding='latin1')
df_superstore.head()
```

```python
df_superstore.columns

```

```python
df_superstore.info
```

```python
df_superstore.describe
```

```python
df_superstore.drop(columns=['Order ID','Row ID','Customer Name','Country','Customer ID'])
```

```python
df_superstore['Order Date'] = pd.to_datetime(df_superstore['Order Date'])
df_superstore['Ship Date'] = pd.to_datetime(df_superstore['Ship Date'])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in ['Category','Sub-Category','State','City','Product Name']:
  df_superstore[col] = le.fit_transform(df_superstore[col])
```

```python
df_superstore.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plots
sns.scatterplot(x='Discount', y='Sales', data=df_superstore)
plt.title("Discount vs Sales")

sns.scatterplot(x='Discount', y='Profit', data=df_superstore)
plt.title("Discount vs Profit")
```
